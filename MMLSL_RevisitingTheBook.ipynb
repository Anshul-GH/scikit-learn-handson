### Types of Machine Learning
---

1. **Machine Learning**:  
The field of study that talks about the ability of the machines to learn without being explicitly programmed to do so.


2. **Supervised Learning**:  
The algorithm learns the behavior from the examples and then make predictions for new input.


3. **Unsupervised Learning**:  
The algorithm does not have any examples or feedback to depend on. It clusters the input records based on observing patterns and grouping similar behavior items together


4. **Semi-supervised Learning**:  
A combination of both supervised and unsupervised learning. Example: Speech Analysis, Protein Sequence Classification, Web Content Classification. 


5. **Reinforcement Learning**:  
Closer to the supervised learning end of the spectrum. They receive a recurring feedback based the actions - a reward for the correct action and penalty for an incorrect one.


---

### Machine Learning Tasks

1. **Classification**:  
The algorithm must learn to predict discrete values for one or more response variables from one or more features. That is the algorithm must predict the most probable class, category or label for new observations.


2. **Regression**:  
The algorithm must predict the value of one or more continuous response variables from one or more features.


3. **Clustering**:  
Assigns the observations to groups such that the observations within one group are more similar to each other than the observations within another group. However, it cannot provide a definite label to the clusters.


4. **Dimensionality Reduction**:  
Process of discovering the features that account for the greatest changes in the response variable. It can also be used to visualize the data.


---

### Types of Datasets

1. **Training Data**:  
Collection of past or existing observations. Contains a set of features or explanatory variables and a set of response variables (or a single variable).

2. **Testing Data**:  
Another dataset, preferably exclusive to the training set, with both explanatory and response variable(s) defined. However, it is used to evaluate the performance of the model using some performance metric.

3. **Validation Data**:  
A validation set is used to tune the hypermarameters for the learning algorithm. Hyperparameters are the variables that control how the algorithm learns from the training data.


---

### Bias and Variance

1. **Bias**:  
    - The model is indifferent to training data
    - Produces the same errors in prediction irrespective of the training dataset used
    - The model **UNDERFITs** the training data


2. **Variance**:  
    - The model is very sensitive to training data
    - Produces different errors in prediction depending on which training dataset is used
    - The model **OVERFITs** the training data

### Important Evaluation Parameters for Classification - Definitions

1. **True Positive**:  
Instances that beloing to class A are also predicted to be belonging to class A.  
There is fire and the alarm rightly rang.

1. **True Negative**:  
Instances that beloing to class B are also predicted to be belonging to class B.  
There is no fire and the alarm rightly did not rang.

1. **False Positive**:  
Instances that beloing to class B are predicted to be belonging to class A. Type I error.  
There is no fire but the alarm rang.

1. **False Negative**:  
Instances that beloing to class A are predicted to be belonging to class B. Type II errror.  
There is fire but the alarm did not rang.
